### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework written in Python that is designed to be lightweight and easy to use. It is called a "micro" framework because it does not require particular tools or libraries and does not impose a specific way of handling web application features. This makes Flask flexible and easy to customize according to the specific needs of the developer.

Some advantages of using Flask for web development include:

1. Easy to learn and use: Flask is straightforward and easy to learn, making it a great choice for beginners. The framework is designed to have a small learning curve, making it quick and easy to get started.

2. Lightweight: Flask is a lightweight framework, which means it does not require much overhead to get started. This makes it ideal for developing small to medium-sized applications.

3. Flexible: Flask provides developers with the freedom to choose their own tools and libraries for various tasks, such as database management and authentication. This makes it easy to customize Flask to suit the specific needs of a project.

4. Extensible: Flask is highly extensible, which means developers can easily add new features or functionalities to their application as needed.

5. Large Community: Flask has a large and active community of developers who contribute to the development of the framework and create plugins, extensions, and add-ons that can be used to enhance the functionality of Flask applications

### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.



```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hello World!!"

if __name__ == "__main__":
    app.run()
```

### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask refers to the mechanism used to associate a URL (or route) with a specific function that will be executed when that URL is requested by a client. The Flask @app.route() decorator is used to define app routes in Flask.

We use app routes in Flask to provide a way for clients to interact with our application. When a client makes a request to a specific URL, Flask uses the routing information we have defined to determine which function should be called to handle the request. The function can then generate a response, which is sent back to the client.

App routing is a fundamental part of web application development in Flask because it enables us to create dynamic, interactive applications that can respond to user input in real-time. It also allows us to organize our code into smaller, more manageable functions, which makes it easier to maintain and update our application over time.

### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: 
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    str = "Company Name: ABC Corporation \n Location: India \n Contact Detail: 999-999-9999"
    return str

if __name__ == '__main__':
    app.run(debug=True)
```

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL to the given endpoint with the method provided. The endpoint is a function that is decorated with the @app.route() decorator.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

@app.route('/user/<username>')
def user_profile(username):
    return f'This is the profile page for {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
        print(url_for('user_profile', username='John Doe'))
```
