### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Histogram Plot: A histogram plot is used to visualize the distribution of a single variable. It displays the frequency of occurrence of values in a given range. It can help identify the shape of the distribution, whether it's skewed or symmetrical, and if there are any outliers.

2. Scatter Plot: A scatter plot is used to display the relationship between two variables. It shows how much one variable is affected by another. It can help identify patterns in the data, such as clusters or outliers.

3. Line Plot: A line plot is used to display trends over time. It is often used to plot time series data or data that is continuous in nature. It can help identify trends, cycles, and patterns in the data.

4. Bar Plot: A bar plot is used to compare the values of different categories. It is often used to display categorical data. It can help identify the distribution of values across categories and can be used to make comparisons between different groups.

5. Heat Map: A heat map is used to display the correlation between two variables. It shows the strength of the relationship between two variables by using colors to represent different levels of correlation. It can help identify patterns in the data and can be used to identify which variables are most strongly correlated.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
data_2 = sns.load_dataset("fmri")
```

```python
sns.lineplot(x = data_2.timepoint, y=data_2.signal);
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
data_3 = sns.load_dataset("titanic")
sns.boxplot(x='pclass', y='age', data=data_3);
```

```python
sns.boxplot(x='pclass', y='fare', data=data_3);
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
data_4 = sns.load_dataset("diamonds")
sns.histplot(data=data_4, x='price',hue='cut');
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


```python
data_5 = sns.load_dataset("iris")
sns.pairplot(data=data_5, hue='species');
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import matplotlib.pyplot as plt
data_6 = sns.load_dataset("flights")
flights_pivot = data_6.pivot("month", "year", "passengers")
fig, ax = plt.subplots(figsize=(16,10))
sns.heatmap(flights_pivot, cmap="YlGnBu",annot=True);
```
