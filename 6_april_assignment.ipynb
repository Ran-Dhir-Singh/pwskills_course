### Q1. What is the mathematical formula for a linear SVM?

A linear SVM (Support Vector Machine) is a binary classifier that tries to find the best linear decision boundary between two classes. The decision boundary is represented by a hyperplane in the feature space.

The mathematical formula for a linear SVM can be expressed as follows:

Given a set of training data {(x1, y1), (x2, y2), ..., (xn, yn)} where xi is a feature vector of length d and yi is its corresponding label (+1 or -1), the goal of a linear SVM is to find a hyperplane w⋅x+b=0 that separates the two classes with maximum margin.

Here, w is a weight vector of length d and b is a scalar bias term. The dot product w⋅x represents the projection of x onto w, and the sign of w⋅x+b determines the predicted class label for x.

The optimal hyperplane is found by solving the following constrained optimization problem:

minimize (1/2) ||w||^2

subject to yi(w⋅xi+b) >= 1 for all i = 1, 2, ..., n


Here, ||w||^2 is the squared Euclidean norm of the weight vector w, which penalizes the complexity of the model and promotes sparsity. The constraint yi(w⋅xi+b) >= 1 ensures that all training examples are correctly classified with a margin of at least 1. This constraint can be viewed as a safety margin that prevents overfitting and improves generalization.

### Q2. What is the objective function of a linear SVM?

The objective function of a linear SVM (Support Vector Machine) is to find the hyperplane that maximizes the margin between the two classes in a binary classification problem. The margin is defined as the distance between the hyperplane and the closest data points from each class.

The objective function of a linear SVM can be expressed as follows:

minimize (1/2) ||w||^2

subject to yi(w⋅xi+b) >= 1 for all i = 1, 2, ..., n

where w is the weight vector that defines the hyperplane, b is the bias term, and xi and yi are the feature vector and the corresponding label of the ith training example, respectively.

The objective function consists of two parts: the regularization term (1/2) ||w||^2 and the classification constraint yi(w⋅xi+b) >= 1. The regularization term penalizes the complexity of the model and promotes sparsity by minimizing the squared Euclidean norm of the weight vector w. The classification constraint ensures that all training examples are correctly classified with a margin of at least 1. This constraint can be viewed as a safety margin that prevents overfitting and improves generalization.

### Q3. What is the kernel trick in SVM?

The kernel trick is a technique used in Support Vector Machines (SVMs) that allows them to efficiently handle non-linearly separable data by mapping the input features into a higher-dimensional space. The kernel trick avoids the explicit computation of the mapping function, which can be computationally expensive or even impossible in some cases.

The basic idea of the kernel trick is to define a kernel function that measures the similarity between pairs of data points in the input space. The kernel function implicitly maps the input data into a higher-dimensional feature space, where it is easier to find a linear decision boundary that separates the two classes.

Mathematically, a kernel function K(x, y) takes two input vectors x and y and returns their inner product in the feature space:

K(x, y) = φ(x)⋅φ(y)

where φ(x) and φ(y) are the corresponding feature vectors in the higher-dimensional space.

Some commonly used kernel functions in SVMs include the linear kernel, the polynomial kernel, the radial basis function (RBF) kernel, and the sigmoid kernel. The choice of kernel function depends on the nature of the data and the problem at hand.

### Q4. What is the role of support vectors in SVM Explain with example

In SVM, support vectors are the data points that are closest to the decision boundary or hyperplane, and they play a critical role in defining the optimal hyperplane and making predictions.

The role of support vectors in SVM can be explained with an example of a binary classification problem with two classes (positive and negative), where the data points are two-dimensional vectors (x1, x2).

Let's assume that we have trained a linear SVM model on a dataset of 100 data points, with 50 in each class. After training, the SVM finds the hyperplane that separates the two classes with maximum margin, and this hyperplane is defined by a weight vector w and a bias term b.

The support vectors are the data points that lie closest to the hyperplane, and they are used to define the margin of the hyperplane. The margin is the distance between the hyperplane and the closest data points from each class.

In this example, let's assume that there are 10 support vectors in the dataset, with 5 in each class. These support vectors are the ones that lie on the margin or within a certain distance from the margin, and they are the ones that contribute to the decision boundary.

When making a prediction on a new data point, the SVM computes the dot product of the weight vector w and the feature vector of the new data point, and adds the bias term b. If the result is positive, the SVM predicts that the new data point belongs to the positive class, and if it is negative, the SVM predicts that the new data point belongs to the negative class.

The support vectors play a critical role in making these predictions because they are the ones that define the hyperplane and the margin, which in turn determines the prediction boundary. The decision boundary is defined by the support vectors, and the location of the new data point with respect to the support vectors determines the class label of the new data point.

### Q6. SVM Implementation through Iris dataset.

* Load the iris dataset from the scikit-learn library and split it into a training set and a testing set
* Train a linear SVM classifier on the training set and predict the labels for the testing set
* Compute the accuracy of the model on the testing set
* Plot the decision boundaries of the trained model using two of the features
* Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

from sklearn.svm import LinearSVC

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data[:, :2], iris.target, test_size=0.3, random_state=42)
```

```python
svm_clf = LinearSVC(C=1, loss="hinge")
svm_clf.fit(X_train, y_train)
```

```python
y_pred = svm_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
plt.figure(figsize=(10, 6))
x_min, x_max = iris.data[:, 0].min() - 0.5, iris.data[:, 0].max() + 0.5
y_min, y_max = iris.data[:, 1].min() - 0.5, iris.data[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("Decision boundary with C=1")
plt.show()
```

```python
for c in [0.01, 0.1, 1, 10]:
    svm_clf = LinearSVC(C=c, loss="hinge")
    svm_clf.fit(X_train, y_train)
    y_pred = svm_clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy with C={}: {}".format(c, accuracy))
```
