    * Advantages:
    
    It is a simple and quick technique to implement.

    * Disadvantages:

3. Hot Deck Imputation:

    * Advantages:

    * Advantages:
    
    It works well when the missing data is not missing at random.
    
    It can preserve the correlation between variables.
    
    * Disadvantages:

    It can introduce bias in the data if the missing data is not missing at random.

    * Advantages:

There are several transformations that can be applied to features to improve normality. Some common transformations include:

1. Logarithmic transformation: This transformation is useful when the data has a long-tailed distribution. Taking the logarithm of the data can help to make the distribution more symmetrical.

2. Square root transformation: This transformation is similar to the logarithmic transformation but is often used when the data contains negative values.

3. Box-Cox transformation: This is a family of transformations that can be used to transform non-normal data into normal data. The Box-Cox transformation is defined as:

y(lambda) = (y^lambda - 1) / lambda if lambda != 0

= log(y) if lambda = 0

where y is the original data and lambda is a parameter that is chosen to maximize the normality of the transformed data.

4. Yeo-Johnson transformation: This is another family of transformations that can be used to transform non-normal data into normal data. The Yeo-Johnson transformation is a modified version of the Box-Cox transformation that can handle both positive and negative data.

5. Power transformation: This transformation involves raising the data to a power that is less than 1. This can help to reduce the variance of the data and make the distribution more normal.
    

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
red_wine_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Separate the features and the target variable
X = red_wine_df.drop('quality', axis=1)
y = red_wine_df['quality']

# Standardize the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# Create a PCA object and fit it to the standardized data
pca = PCA()
pca.fit(X_std)
```

```python
import matplotlib.pyplot as plt

# Plot the cumulative explained variance ratio as a function of the number of principal components
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```
