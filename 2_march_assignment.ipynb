### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

`Matplotlib` is a popular Python library used for creating data visualizations. It provides a variety of plotting options and customization features that enable users to create high-quality and informative graphics.

`Matplotlib` is used to create a wide range of plots such as line plots, scatter plots, bar plots, histograms, and more. It can be used for exploratory data analysis, scientific visualization, and presentation of data in reports and publications.

The Pyplot module is a sublibrary of Matplotlib that provides a simple interface for creating plots. Some common plots that can be created using Pyplot are:

1. Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. It is useful for showing trends and patterns in data over time.

2. Scatter Plot: A scatter plot is a graph that displays data as a collection of points. It is useful for showing the relationship between two variables.

3. Bar Plot: A bar plot is a graph that displays data as a series of rectangular bars. It is useful for comparing data across categories.

4. Histogram: A histogram is a graph that displays the distribution of data. It is useful for showing the frequency of data values in a dataset.

5. Pie Chart: A pie chart is a graph that displays data as a circle divided into segments. It is useful for showing the relative proportions of different categories of data.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title of the plot");
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

The `subplot()` function in Matplotlib is used to create multiple plots in a single figure. It takes three arguments: the number of rows, the number of columns, and the index of the subplot to create. The index starts at 1 and increases from left to right, top to bottom.

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(x1,y1)
ax[0,0].set_title("Line1")

ax[0,1].plot(x2,y2)
ax[0,1].set_title("Line2")

ax[1,0].plot(x3,y3)
ax[1,0].set_title("Line3")

ax[1,1].plot(x4,y4)
ax[1,1].set_title("Line4")

fig.suptitle("Subplots")

plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot, also known as a bar chart, is a graphical representation of categorical data with rectangular bars of varying heights or lengths. Bar plots are used to show the comparison of different categories of data.

Bar plots are useful for displaying information that can be divided into discrete categories. They are especially useful when comparing the size or quantity of each category. They allow for a quick and easy visual comparison of the data and are widely used in many different fields, including business, finance, and scientific research.

In a bar plot, the height or length of each bar represents the quantity or size of each category. The bars can be oriented vertically or horizontally, and multiple bars can be grouped together for easier comparison. Bar plots can also be stacked or overlaid with other bar plots to show more complex relationships between the data.

Overall, bar plots are an effective tool for visualizing categorical data and making meaningful comparisons between different categories.

```python
plt.bar(company, profit);
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data. It shows the median, quartiles, and outliers of the data.

Box plots are used to visualize the spread and skewness of the data. They provide a quick and easy way to identify the range and distribution of the data, as well as any outliers that may be present.

In a box plot, a box is drawn from the first quartile (Q1) to the third quartile (Q3), with a line drawn at the median. Whiskers are drawn from the edges of the box to the minimum and maximum values within 1.5 times the interquartile range (IQR) from the edges of the box. Any data points outside this range are considered outliers and are plotted individually.

```python
data = [box1, box2]

fig, ax = plt.subplots()

# create box plot
ax.boxplot(data)

# set title and labels
ax.set_title('Box plot')
ax.set_xlabel('Data')
ax.set_ylabel('Value')

plt.show()
```
